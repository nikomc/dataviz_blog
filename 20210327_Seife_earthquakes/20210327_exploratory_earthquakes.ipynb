```python
import numpy as np
import math
import pandas as pd
import folium
```

```python
df = pd.read_csv('earthquakes.tsv', sep='\t')
```

```python
df.head()
```

```python
df.max()
```

```python
df.info
```

```python
map_1 = folium.Map(location=[40, -110], zoom_start=2.5)
for country in pd.unique(df["id"]):
    #Pass only values of data, not dataframe
    lon_lat_coun = df.loc[df["id"] == country,["latitude", "longitude", "date", "location", "magnitude"]].values
    for latitude, longitude, date, location, magnitude in lon_lat_coun:
        folium.Circle(
            radius=float(math.sqrt(magnitude))*20000,
            location=[latitude, longitude],
            popup='Date: {}, Mag: {}, Loc: {}'.format(date, magnitude, location),
            color="crimson",
            fill=True,
            ).add_to(map_1)
#Uncomment only if you want to save file map        
map_1.save('earthquake_data.html')
```
